```python
import pandas as pd
from bs4 import BeautifulSoup

# Load the CSV file into a dataframe
df_phish = pd.read_csv("enron_spam.csv")
df_ham = pd.read_csv("enron_ham.csv", encoding='ISO-8859-1')
df_phish['label'] = 1
df_ham['label'] = 0
# concatenate dataframes vertically
df_concat = pd.concat([df_phish, df_ham])

# randomize rows
df = df_concat.sample(frac=1, random_state=42)
df.isna().sum()
```

```python
# drop rows with missing values in column A
df = df.dropna(subset=['Body'])
```

```python
df.isna().sum()
```

```python
import re

for i, row in df.iterrows():
    body = row['Body']
    if pd.isna(body):
        continue
    if re.match(r'<html>', body):
        df.at[i, 'is_html'] = True
```

```python
def extract_urls(html):
    soup = BeautifulSoup(html, 'html.parser')
    urls = []
    for link in soup.find_all('a', href=True):
        urls.append(link['href'])
    for tag in soup.find_all(href=re.compile("^http")):
        urls.append(tag.get('href'))
    print(urls[0]) if len(urls) > 0 else print(urls)
    return urls
```

```python
df.dropna(subset=['Body'], inplace=True)
```

```python
import pandas as pd
from bs4 import BeautifulSoup
import re

url_regex = r"(?P<url>https?://[^\s]+)"

# Load the data

# Replace NaN values with empty strings
#df['Body'] = df['Body'].fillna('')

# Apply the function to each row in the DataFrame
df['html_urls'] = df['Body'].apply(extract_urls)

# Display the results
print(df.head())
```

```python


# Create a new column for the body text
df["body_text"] = ""

# Loop through each row in the dataframe
for i, row in df.iterrows():
    # Check if the body is HTML
    if row["is_html"]:
        # Parse the HTML with BeautifulSoup
        soup = BeautifulSoup(row["Body"], "html.parser")
        # Extract the text from the HTML
        body_text = soup.get_text()
    else:
        # If the body is not HTML, use the raw text
        body_text = row["Body"]
    # Store the extracted text in the new column
    df.at[i, "body_text"] = body_text
```

```python

# Define a regular expression to match URLs
url_regex = r'https?://(?:[-\w.]|(?:%[\da-fA-F]{2}))+'

# Extract URLs from the HTML using the regular expression
df['body_urls'] = df['body_text'].apply(lambda x: re.findall(url_regex, str(x)))

```

```python
df
```

```python

email_regex = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'

def replace_emails(body):
    return re.sub(email_regex, 'email', str(body))

def replace_url(body):
    return re.sub(url_regex, 'URL', str(body))

df['body_text'] = df['body_text'].apply(replace_emails)

df['body_text'] = df['body_text'].apply(replace_url)
```

```python
def clean_text(text):
    # convert to lowercase
    text = text.lower()
    # remove HTML tags
    text = re.sub('<[^<]+?>', '', text)
    # replace contractions
    text = re.sub(r"won\'t", "will not", text)
    text = re.sub(r"can\'t", "can not", text)
    text = re.sub(r"n\'t", " not", text)
    text = re.sub(r"\'re", " are", text)
    text = re.sub(r"\'s", " is", text)
    text = re.sub(r"\'d", " would", text)
    text = re.sub(r"\'ll", " will", text)
    text = re.sub(r"\'t", " not", text)
    text = re.sub(r"\'ve", " have", text)
    text = re.sub(r"\'m", " am", text)
    text = re.sub(r'\b\w{14,}\b', '', text)
    # remove URLs
    text = re.sub(r'http\S+', '', text)
    # replace email addresses
    text = re.sub(r'\S+@\S+', 'email', text)
    # remove numbers
    text = re.sub(r'\d+', 'number', text)
    # remove extra whitespace
    text = re.sub(r'\s+', ' ', text)
    return text
```

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt') # download the tokenizer
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
from nltk.stem import PorterStemmer

porter_stemmer = PorterStemmer()
#df['text'] = df.apply(lambda row: print(row +' here it is ------------------------------------'))
#print(df.Subject)
df['text'] = df.apply(lambda row: row['Subject'] + ' ' + row['body_text'],axis=1)
df['text'] = df.apply(lambda row: clean_text(row['text']),axis=1)
df['tokenized_text'] = df.apply(lambda row: word_tokenize(row['text']),axis=1)
df['tokenized_text'] = df.apply(lambda row: [word for word in row['tokenized_text'] if word not in stop_words],axis=1)
df['tokenized_text'] = df.apply(lambda row: [porter_stemmer.stem(word) for word in row['tokenized_text']],axis=1)
df['text'] = df.apply(lambda row: ' '.join(row['tokenized_text']),axis=1)
df.head()
```
