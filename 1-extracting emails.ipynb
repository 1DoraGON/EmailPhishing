```python
import os
import csv
import email
from email.header import decode_header


# Set the path to the main directory containing the email subdirectories
main_dir = "./enron labeled/compressed/rows original distribution ham/"

# Create a list to store the extracted information
data = []
def get_field(msg, field):
    value = msg.get(field)
    if value is None:
        return "Nan"
    else:
        return decode_header(value)[0][0]

# Loop through each subdirectory in the main directory
for subdir in os.listdir(main_dir):
    # Create the full path to the subdirectory
    path = os.path.join(main_dir, subdir)
    # Check if the path is a directory
    if os.path.isdir(path):
        # Loop through each email file in the subdirectory
            for subsubdir in os.listdir(path):
                subpath = os.path.join(path, subsubdir)
                if os.path.isdir(subpath):
                    for filename in os.listdir(subpath):
                        
                        try:
                            with open(os.path.join(subpath, filename), "r") as f:
                                #print(subpath)
                                # Parse the email using the email module
                                msg = email.message_from_file(f)
                                # Extract the desired fields from the email
                                message_id = get_field(msg, "Message-ID")
                                date = get_field(msg, "Date")
                                sender = get_field(msg, "From")
                                recipient = get_field(msg, "To")
                                subject = get_field(msg, "Subject")
                                body = msg.get_payload()

                                # Append the extracted information to the data list
                                data.append([message_id, date, sender, recipient, subject, body])
                                #print([message_id, date, sender, recipient, subject, body])
                        except PermissionError:
                            print(f"Skipping {filename}: Permission denied.")
                            continue

# Write the data to a CSV file
with open("enron_ham.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Message ID", "Date", "Sender", "Recipient", "Subject", "Body"])
    writer.writerows(data)
```

```python
import csv
import re
import os

# set the directory path where the email files are stored
directory_path = './enron/enron1/spam/sample'

# define a regular expression pattern to match the subject line
subject_pattern = re.compile(r'Subject: (.*)')

# create a CSV file to store the subject and body
csv_file = open('emails.csv', mode='w', newline='')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['Subject', 'Body'])

# loop through each email file in the directory
for filename in os.listdir(directory_path):
    if filename.endswith('.txt'):
        print('hi')
        email_file = open(os.path.join(directory_path, filename), mode='r', encoding='utf-8')
        email_contents = email_file.read()
        print(email_contents)
        email_file.close()

        # extract the subject and body from the email contents
        subject_match = subject_pattern.search(email_contents)
        if subject_match:
            subject = subject_match.group(1)
            body = email_contents.replace(subject_match.group(0), '', 1).strip()

            # write the subject and body to the CSV file
            csv_writer.writerow([subject, body])

# close the CSV file
csv_file.close()
```
